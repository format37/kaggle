```python
def should_hit(player_total, dealer_card_val, player_aces):
    
    global policy
    global choice
    global obs_player_total
    global obs_dealer_card_val
    global obs_player_aces    
    
    obs_player_total=player_total
    obs_dealer_card_val=dealer_card_val
    obs_player_aces=player_aces
    
    if policy[player_total][dealer_card_val][player_aces]==0.5:
        #choice = True if randint(0, 1)==1 else False
        choice=False
    if policy[player_total][dealer_card_val][player_aces]<0.5:
        choice = False
    else:
        choice = True
    
    return choice
```

```python
def learning_rate_report(policy):
    positive = 0
    neutral = 0
    negative = 0
    for p in range(22):
        for d in range(22):
            for a in range(6):
                if policy[p][d][a]<0.5:
                    negative += 1
                elif policy[p][d][a]==0.5:
                    neutral += 1
                elif policy[p][d][a]>0.5:
                    positive += 1
    return negative,neutral,positive
```

```python
def plot_policy():
    global policy
    policy_crop=policy.reshape(22,22*12)
    my_list = [list(row) for row in policy_crop]
    trace = go.Heatmap(z=my_list)
    data=[trace]
    plot_file_name='basic-heatmap.html'
    plotly.offline.plot(data, filename=plot_file_name)
    return plot_file_name
```
